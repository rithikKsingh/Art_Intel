```python
### Textloader

from langchain_community.document_loaders import TextLoader

loader=TextLoader("../data/text_files/spring_boot_guide.txt",encoding="utf-8")
document=loader.load()
print(document)

# Textloader by default give data in structured format
#  which contains metadata: source and page content 
```

```python
### Directory loader

from langchain_community.document_loaders import DirectoryLoader

## load all the text files from the directory
dlr_loader=DirectoryLoader(
    "../data/text_files",
    glob="**/*.txt", ## Pattern to match all the files
    loader_cls=TextLoader, ## here we are using text loader as text data is getting read.
    #If you are reading pdf, you can use PDF
    loader_kwargs={'encoding':'utf-8'},
    show_progress=False
)

document=dlr_loader.load()
print(document)
```

```python
### Reading PDFs

from langchain_community.document_loaders import PyPDFLoader,PyMuPDFLoader

## load all the text files from the directory
dlr_loader=DirectoryLoader(
    "../data/pdf",
    glob="**/*.pdf", ## Pattern to match all the files
    loader_cls=PyMuPDFLoader, 
    # loader_kwargs={'encoding':'utf-8'}, #here no need to apply encoding
    show_progress=False
)

pdf_documents=dlr_loader.load()
pdf_documents
```
